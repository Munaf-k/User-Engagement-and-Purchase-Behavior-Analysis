```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
df = pd.read_csv('marketing_data new.csv')

df.head()
```

```python
# Checking data types
df.info()
```

```python
# Converting to numeric datatype
columns = [
    'Likes', 
    'Shares', 
    'Comments', 
    'Clicks', 
    'Engagement_with_Ads', 
    'Time_Spent_on_Platform', 
    'buy'
]

for column in columns:
    df[column] = df[column].str.replace(',', '').astype(float)
```

```python
df.info()
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Statistical summary of dataset
df.describe()
```

```python
print(df['Engagement_Level'].value_counts())
```

```python
print(df['Purchase_Likelihood'].value_counts())
```

```python
# Plotting purchase likelihood
df['Purchase_Likelihood'].value_counts().plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Purchase Likelihood Distribution')
plt.xlabel('Purchase Likelihood (0 = Unlikely, 1 = Likely)')
plt.ylabel('Count')
plt.show()
```

```python
numeric_columns = ['Likes', 'Shares', 'Comments', 'Clicks', 'Engagement_with_Ads', 'Time_Spent_on_Platform']

# Plotting histograms
df[numeric_columns].hist(figsize=(10, 8), bins=10, color='skyblue')
plt.suptitle('Engagement Metrics Distribution')
plt.tight_layout()
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_columns])
plt.title('Boxplot of Engagement Metrics')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[numeric_columns].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Grouping by Engagement_Level and calculating mean
engagement_summary = df.groupby('Engagement_Level')[numeric_columns].mean()
print(engagement_summary)
```

```python
engagement_summary.plot(kind='bar', figsize=(10, 6))
plt.title('Average Engagement Metrics by Engagement Level')
plt.ylabel('Average Value')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Engagement_Level', y='Time_Spent_on_Platform', data=df)
plt.title('Time Spent on Platform by Engagement Level')
plt.show()
```

```python
# Saving data to a new CSV fil
df.to_csv('cleaned_marketing_data.csv', index=False)
```
